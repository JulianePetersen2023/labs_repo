```python
def top10 (df):
    df.rename(columns={'Unnamed: 0': 'subject_ID'}, inplace=True)
    df1 = pd.concat([df.subject_ID, df.subject, df.score], axis=1)
    df_sorted = df1.sort_values(by=["score", "subject_ID"], ascending=[False, True])
    df = df_sorted.head(10)
    return(df)
```

```python
def highest_average_solutions(df: pd.DataFrame) -> int:
    """ Returns the solutions id with the highest average score """
    con1 = df.solutions==1
    con2 = df.solutions==2
    con3 = df.solutions==3
    df1 =df[con1]
    df2 =df[con2]
    df3 =df[con3]
    if df1.score.mean() > df2.score.mean() and df1.score.mean() > df3.score.mean():
        y=1
    elif df2.score.mean() > df1.score.mean() and df2.score.mean() > df3.score.mean():
        y=2
    else:
        y=3
    solution_id= y
    return solution_id
```
