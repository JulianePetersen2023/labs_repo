```python
import pandas as pd 

df = pd.read_csv("https://raw.githubusercontent.com/Esther-Velasco/esther_labs/main/W3/Mini%20project/Startups%20in%202021%20end.csv", on_bad_lines='skip')
df
```

```python
df.nunique()
```

```python
df.Industry.value_counts()
```

```python
industry_mapping = {
    "Artificial intelligence":"Artificial intelligence",
    "Artificial Intelligence":"Artificial intelligence"                    
}

df['Industry']= list(map(lambda x:industry_mapping.get(x,x),df['Industry']))
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns = df.columns.str.lower()
```

```python
startups = df.drop(("select_investors"), axis = 1)
startups = df.drop(("unnamed:_0"), axis = 1)
startups
```

```python
startups.country.value_counts()
```

```python
country_mapping = {
    "United States":"United States",
    "United States,":"United States",
    "Indonesia":"Indonesia",
    "Indonesia,":"Indonesia"
}

startups['country']= list(map(lambda x:country_mapping.get(x,x),startups['country']))
```

```python
startups.country.unique()
```

```python
city_mapping = {
    "South San Francisco":"San Francisco",
    "Tel Aviv-Yafo,":"Tel Aviv",
    "Washington":"Washington DC",
}

startups['city']= list(map(lambda x:city_mapping.get(x,x),startups['city']))
```

```python
startups
```

```python
df_clean = startups.drop(("select_investors"), axis = 1)

#we figured we needed to create the foreign key values data so we create index columns for city_id, 
#country_id, startup_id and industry_id
```

```python

df_clean["country_id"]= range(1, len(df_clean) + 1)
df_clean["city_id"] = range(1, len(df_clean) + 1)
df_clean["startup_id"] = range(1, len(df_clean) + 1)
df_clean["industry_id"] = range(1, len(df_clean) + 1)
df_clean
```

```python
# Specify the path where you want to save the CSV file
output_path = "https://github.com/Esther-Velasco/esther_labs/tree/main/W3/Mini%20project"

# Use the to_csv method to save the DataFrame to a CSV file
df_clean.to_csv("startups_index.csv", index=False)

```
