### Data Cleaning

```python
bs = pd.read_csv('bullying_2018.csv', sep= ";")
```

```python
### to make female 1 and male 0

bs['Sex'] = bs['Sex'].replace({'Female': 1, 'Male': 0})
bs['Sex'] 
```

```python
### to extract the digits

bs['Custom_Age'] = bs['Custom_Age'].str.extract('(\d+)')
bs['Custom_Age']
```

```python
### to extract the digits
bs['Physically_attacked'] = bs['Physically_attacked'].str.extract('(\d+)')
bs['Physically_attacked']
```

```python
### Extract digit groups or ranges
bs['Physical_fighting'] = bs['Physical_fighting'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')

# Function to process the extracted groups
def process_groups(groups):
    result = []
    for group in groups:
        if group[1]:  # If there is a range
            result.extend(map(str, range(int(group[0]), int(group[1]) + 1)))
        else:
            result.append(group[0])
    return ', '.join(result)

# Apply the processing function to the extracted groups
bs['Physical_fighting'] = bs['Physical_fighting'].apply(process_groups)

# Display the result
print(bs['Physical_fighting'])
```

```python
### Extract digit groups or ranges
bs['Close_friends'] = bs['Close_friends'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')

# Function to process the extracted groups
def process_groups(groups):
    result = []
    for group in groups:
        if group[1]:  # If there is a range
            result.extend(map(str, range(int(group[0]), int(group[1]) + 1)))
        else:
            result.append(group[0])
    return ', '.join(result)

# Apply the processing function to the extracted groups
bs['Close_friends'] = bs['Close_friends'].apply(process_groups)

# Display the result
bs['Close_friends'].head()
```

```python
### Extract digit groups or ranges
bs['Miss_school_no_permission'] = bs['Miss_school_no_permission'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')

# Function to process the extracted groups
def process_groups(groups):
    result = []
    for group in groups:
        if group[1]:  # If there is a range
            result.extend(map(str, range(int(group[0]), int(group[1]) + 1)))
        else:
            result.append(group[0])
    return ', '.join(result)

# Apply the processing function to the extracted groups
bs['Miss_school_no_permission'] = bs['Miss_school_no_permission'].apply(process_groups)

# Display the result
print(bs['Miss_school_no_permission'])
```

```python
### Get mean
# Extract digit groups or ranges
bs['Physical_fighting'] = bs['Physical_fighting'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')

# Function to calculate the mean of extracted numbers
def calculate_mean(groups):
    numbers = []
    for group in groups:
        if group[1]:  # If there is a range
            numbers.extend(range(int(group[0]), int(group[1]) + 1))
        else:
            numbers.append(int(group[0]))
    return sum(numbers) / len(numbers) if numbers else None

# Calculate mean for each row
bs['Mean_Physical_fighting'] = bs['Physical_fighting'].apply(calculate_mean)

# Display the result
print(bs['Mean_Physical_fighting'])
```

```python
### Get mean
# Extract digit groups or ranges
bs['Miss_school_no_permission'] = bs['Miss_school_no_permission'].str.findall(r'(\d+)(?:\s*-\s*(\d+))?')

# Function to calculate the mean of extracted numbers
def calculate_mean(groups):
    numbers = []
    for group in groups:
        if group[1]:  # If there is a range
            numbers.extend(range(int(group[0]), int(group[1]) + 1))
        else:
            numbers.append(int(group[0]))
    return sum(numbers) / len(numbers) if numbers else None

# Calculate mean for each row
bs['Mean_school_absence'] = bs['Miss_school_no_permission'].apply(calculate_mean)

# Display the result
print(bs['Mean_school_absence'])
```
