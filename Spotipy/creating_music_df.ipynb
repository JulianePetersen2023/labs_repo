```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB") #username commonly spotify
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])  #extend adds this list to the other
    return tracks
```
